```python
from langchain_ollama import ChatOllama

llm = ChatOllama(
    model="llama3.1",
    temperature=0,
    # other params...
)
```

```python
from langchain_core.prompts import ChatPromptTemplate

system_template = "Translate the following from English into {language}"

prompt_template = ChatPromptTemplate.from_messages(
    [("system", system_template), ("user", "{text}")]
)
```

```python
prompt = prompt_template.invoke({"language": "tamil", "text": "Are you READY???"})

prompt
```

```python
llm.invoke(prompt)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template("Tell me a Facts about {topic}")

prompt = prompt_template.invoke({"topic": "BMW Cars"})

response = llm.invoke(prompt)

response.content
```

```python
car_models_template = PromptTemplate.from_template("Suggest me Car Models in budget {budget} and of Company {Company}")
car_model_prompt = car_models_template.invoke({"budget": "70000 euros", "Company": "BMW"})
response = llm.invoke(car_model_prompt)
response.content
```

```python
print(response.content)
```

```python
from langchain_core.output_parsers import StrOutputParser
car_models_chain = car_models_template | llm | StrOutputParser()
car_models_chain.invoke({"budget": "70000 euros", "Company": "BMW"})
```

```python
one_word_prompt = PromptTemplate.from_template("Show me the best Car Model in this Range {car_models}")
```

```python
from langchain_core.runnables import RunnableLambda
model_name_step = RunnableLambda(lambda x : print(x))

modified_chain = {"car_models": car_models_chain} | model_name_step | one_word_prompt | llm | StrOutputParser()
```

```python
from langchain_core.runnables import RunnableSequence

modified_chain = RunnableSequence({"car_models": car_models_chain}, model_name_step, one_word_prompt, llm, StrOutputParser())
```

```python
range = modified_chain.invoke({"budget": "70000 euros", "Company": "BMW"})
```

```python
from langchain_core.runnables import RunnableParallel

model1_chain = ChatPromptTemplate.from_template("Show me budget Car model for BMW in the range {range} of 60k euros - 70k euros ") | llm | StrOutputParser()
model2_chain = ChatPromptTemplate.from_template("Show me budget Car model for Toyota in the range {range} of 40k euros - 50k euros ") | llm | StrOutputParser()

Car_parallel = RunnableParallel(model1 = model1_chain, model2 = model2_chain)

Car_parallel.invoke({"range" : range})                                
```
